### Homework 2.3 - 2D Temperature Distribution

Solve 2D heat equation for the following initial conditions:

$$\begin{array}{l}
u(x,0) = \sin (2\pi x)\\
u(x,1) = \sin (2\pi x)\\
u(0,y) = \sin (3\pi y)\\
u(1,y) = \sin (\pi y)
\end{array}$$

where $u(x, y)$ represent the steady state temperature distribution on 2D plane.
 
Plot the 2D heat map of the solution.

```python
D = 100 + 1
u = np.zeros((D, D))

x = np.linspace(0, 1, D)
u[:, 0] = np.sin(2*np.pi*x)
u[:, D-1] = np.sin(2*np.pi*x)

y = np.linspace(0, 1, D)
u[0, :] = np.sin(3*np.pi*y)
u[D-1, :] = np.sin(np.pi*y)

n = 0
while n < 1000:
    difference = 0
    u_old = u.copy()
    for i in range(1, D-1):
        for j in range(1, D-1):
            u[i,j] = (1/4)*(u_old[i+1,j] + u_old[i-1,j] + u_old[i,j+1] + u_old[i,j-1])
            difference += abs(u[i,j] - u_old[i,j])
    difference /= D*D
    if difference < 1e-5:
        break
    n += 1
    
plt.imshow(u, cmap='hot')
plt.show()
```
